## Part III: Functions

### 1. Write a function to convert temperature from kelvin to celsius

```python
def temp_conversion_to_celsius(kelvin):
    """ Temperature conversion from kelvin to celsius """
    return kelvin - 273.15
```

```python
def temp_conversion_to_kelvin(celsius):
    """ Temperature conversion from celsius to kelvin """
    return celsius + 273.15
```

### 2. Write a function to convert temperature to fahrenheit

```python
def temp_conversion_to_fahrenheit(temp, unit):
    """ Temperature conversion from celsius or kelvin to fahrenheit """
    if unit == 'celsius':
        return (temp * 9/5) + 32
    elif unit == 'kelvin':
        return ((temp - 273.15) * 9/5) + 32
    else:
        print("Error: Unit must be celsius or kelvin")
```

### 4) Now write a function that converts from fahrenheit

```python
def temp_conversion_from_fahrenheit(temp, unit):
    """ Conversion from fahrenheit to either celsius or kelvin """
    if unit == 'celsius':
        return (temp - 32) * 5/9
    elif unit == 'kelvin':
        return ((temp - 32) * 5/9) + 273.15
    else:
        print("Error: Unit must be celsius or kelvin")
```

```python
temp_conversion_from_fahrenheit(68, 'celsius')
```

```python
temp_conversion_from_fahrenheit(68, 'kelvin')
```

```python
temp_conversion_from_fahrenheit(70, 'fahrenheit')
```

### 5) Write a function that takes two arguments (feet and inches) and returns height in meters

```python
def meters_conversion_from_feet_and_inches (feet, inches):
    """ Conversion from feet and inches to meters """
    return round((feet / 3.281) + (inches / 39.37), 2)
```

### 6. Write a function takes one argument (height in meters) and returns two arguments (feet and inches)

```python
def feet_and_inches_conversion_from_meters (meters):
    """ Conversion from meters to feet and inches """
    total_inches = meters * 39.37
    feet = int(total_inches // 12)
    inches = round(total_inches % 12)
    if inches == 12:
        feet += 1
        inches = 0
    return feet, inches
```
